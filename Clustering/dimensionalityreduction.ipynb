```python
# Data from scikit-learn's bundled Digits dataset
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# TSNE estimator for dimensionality reduction
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=24)
reduced_data = tsne.fit_transform(digits.data)
reduced_data.shape
```

```python
# Visualize reduced data
import matplotlib.pyplot as plt
scatterplot = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], 
c=digits.target, cmap=plt.cm.get_cmap('Set3_r', 10))
colourbar = plt.colorbar(scatterplot)
```
